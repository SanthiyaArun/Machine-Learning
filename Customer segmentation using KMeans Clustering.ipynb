# Importing the dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

# Data collection and analysis

```python
customer_data=pd.read_csv("Mall_Customers.csv")
```

```python
customer_data.head()
```

```python
customer_data.info()
```

```python
customer_data.isnull().sum()
```

# Choosing the Annual Income Column & Spending Score column

```python
x=customer_data.iloc[:,[3,4]].values
```

```python
x
```

# Choosing the number of clusters

```python
#within cluster sum of square
wcss=[]

for i in range(1,11):
    model=KMeans(i)
    model.fit(x)
    wcss.append(model.inertia_)
wcss
```

```python
#elbow graph

sns.set()
plt.plot(range(1,11),wcss)
plt.title("The Elbow Point Graph")
plt.xlabel("The Number of Clusters")
plt.ylabel("WCSS")
plt.show
```

# Optimum Number of Clusters = 5

# Training the k-Means Clustering Model

```python
model=KMeans(5)
```

```python
model.fit(x)
```

```python
model.fit_predict(x)
```

```python
Y=model.fit_predict(x)
```

```python
customer_data["Cluster"]=Y
customer_data
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[Y==0,0], x[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(x[Y==1,0], x[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(x[Y==2,0], x[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(x[Y==3,0], x[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(x[Y==4,0], x[Y==4,1], s=50, c='blue', label='Cluster 5')

# plot the centroids
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=100, c='cyan', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(customer_data['Annual Income (k$)'], customer_data['Spending Score (1-100)'],c=customer_data['Cluster'],cmap='rainbow')
# plot the centroids
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=100, c='black', label='Centroids')
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a figure with subplots for all five clusters
fig, axs = plt.subplots(1, 5, figsize=(15, 3), sharey=True)

# Loop over the unique values of Cluster column
for i, cluster in enumerate(customer_data['Cluster'].unique()):
    # Filter the data of the current cluster
    cluster_data = customer_data[customer_data['Cluster'] == cluster]
    
    # Extract the values of age and spending score
    age = cluster_data['Age']
    score = cluster_data['Spending Score (1-100)']
    
    # Create a scatter plot of age vs spending score in the current subplot
    axs[i].scatter(age, score)
    axs[i].set_xlabel('Age')
    axs[i].set_title(f'Cluster {cluster}')
    
# Add a y-axis label to the first subplot
axs[0].set_ylabel('Spending Score (1-100)')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group the data by Cluster and count the number of data points in each group
cluster_counts = customer_data.groupby('Cluster').size()

# Print the total number of data points in each cluster
for cluster, count in cluster_counts.iteritems():
    print(f'Cluster {cluster}: {count} data points')
```

```python
# Create a figure with subplots for all five clusters
fig, axs = plt.subplots(1, 5, figsize=(15, 3), sharey=True)

# Loop over the unique values of Cluster column
for i, cluster in enumerate(customer_data['Cluster'].unique()):
    # Filter the data of the current cluster
    cluster_data = customer_data[customer_data['Cluster'] == cluster]
    
    # Extract the values of age and spending score
    age = cluster_data['Annual Income (k$)']
    score = cluster_data['Spending Score (1-100)']
    
    # Create a scatter plot of age vs spending score in the current subplot
    axs[i].scatter(age, score)
    axs[i].set_xlabel('Annual Income (k$)')
    axs[i].set_title(f'Cluster {cluster}')
    
# Add a y-axis label to the first subplot
axs[0].set_ylabel('Spending Score (1-100)')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
